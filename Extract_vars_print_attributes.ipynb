##### (1) To select from s SAS data set a subset of variables that share a common suffix, some of which may have varying middle parts in their names

* Create a macro variable whose value is a horizontal list of variables, with each variable name in double quotes and separated by a comma (PROC SQL into ... from dictionary.columns).

##### (2) To print the attributes (name, type, and attributes) of those variables

* Print name, type, and label of those variables (PROC SQL  from dictionary.columns)

```sas
options nocenter nonotes nodate nosource;
ods html close;
* Create a macro variable that holds a horizontal list of variables of interest;
proc sql noprint;
select quote(trim(name))  into :varlist separated by ', ' 
 from dictionary.columns
where libname= "PUFMEPS" and memname = "H201"
 and (name like "INSURCP%" or name like "MCR%" 
       or name like "MCD__%" 
       or name like "MCD__%"
       or name like "PRX__%"
       or name like "PUB__%"
       or name like "MCREV%"
       or name like "MCDEV%"
       or name like "PRVEV%" 
       or name like "TRIEV%"   
       or name like "MCRPHO%"   
       or name like "MCRPD%"   
       or name like "VARSTR"
       or name like "VARPSU"
       or name like "PERWT%" 
       or name like "agelast");
       %let num_of_variables = &sqlobs;
quit;
%put &=varlist  &=num_of_variables;

```

```sas
* Print the meta data for the variables of interest;
libname pufmeps 'C:\Data';
proc sql ;
select name, type, label
  from dictionary.columns
 where libname= "PUFMEPS" and memname = "H201"
 and name in (&varlist);
quit;
```
